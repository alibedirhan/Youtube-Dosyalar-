<h3><li>man komutu</li></h3>

man komutu "manuel" den gelir ana amacı başka bir komutun ne işe yaradığı hakkında bilgi



<h3><li>ls komutu</li></h3>

ls komutu "liste" den gelir. Bulunduğunuz dizindeki dosyaları terminal ekranında listeler. ls komutunun bazı parametreleri var bunlardan en çok kullanılanı;

<b>ls -a</b>

Bu komut bulunduğunuz dizin içindeki gizli dosyaları gösterir. Nokta ile başlayan klasörler gizli dosyalardır.

<b>ls -l</b>

Bu komut dizin içinde bulunan dosyaların genel özelliklerini gösterir. İzinler, boyut, sahip gibi detaylı bilgilerle dosyaları ve dizinleri listeleyecektir




<h3><li>cd komutu</li></h3>

Bu komut "dizin değiştirme"(change directory) anlamına gelir. Dizinler arası geçişi sağlar ve parametreler barındırır.

<b>cd .. </b>

Bu komut bulunduğunuz dizinden bir önceki dizine geçiş yapmayı sağlar
    




<h3><li>pwd komutu</li></h3>

Bu komut "Çalışan Dizini Göster"(Print Working Directory) hangi dizinde olduğunuzu size söyler. Dizinin(klasörün) kesin ve tam bilgisine ulaştırır



<h3><li>mkdir komutu</li></h3>


mkdir komutu "Dizin oluşturma"(make directory) anlamına gelir. Bir Dizin/Klasör oluşturmak için kullanılır.
